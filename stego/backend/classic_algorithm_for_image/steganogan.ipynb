```python
import numpy as np
import cv2
from skimage.measure import shannon_entropy

# Загрузка изображения
image = cv2.imread('/mnt/0.png', cv2.IMREAD_GRAYSCALE)

# Функция для вычисления локальной энтропии
def local_entropy(image, window_size=3):
    height, width = image.shape
    entropy_map = np.zeros((height, width))

    # Определяем полуширину окна
    half_window = window_size // 2
    for i in range(half_window, height - half_window):
        for j in range(half_window, width - half_window):
            window = image[i-half_window:i+half_window+1, j-half_window:j+half_window+1]
            entropy_map[i, j] = shannon_entropy(window)
    
    return entropy_map

# Получение локальной энтропии
entropy_map = local_entropy(image, window_size=3)

# Визуализация локальной энтропии
import matplotlib.pyplot as plt
plt.imshow(entropy_map, cmap='hot')
plt.colorbar()
plt.title('Локальная энтропия изображения')
plt.show()
```
